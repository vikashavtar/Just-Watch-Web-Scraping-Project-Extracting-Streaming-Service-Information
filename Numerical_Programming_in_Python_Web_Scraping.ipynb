## **Scrapping Movie Title**

````python
# Write Your Code here
import requests
from bs4 import BeautifulSoup

# Base URL of JustWatch
base_url = 'https://www.justwatch.com'

# Replace 'initial_url' with the initial URL you want to scrape
initial_url = base_url + '/in/movies?release_year_from=2000'

# Headers to mimic a browser visit
headers = {'User-Agent': 'Mozilla/5.0'}

# Send an HTTP GET request to the initial URL and parse the HTML content
initial_page = requests.get(initial_url, headers=headers)
soup = BeautifulSoup(initial_page.text, 'html.parser')

# Initialize a list to store movie titles
movie_titles = []

# Scrape for movie URLs and extract titles
for link in soup.find_all('a', class_='title-list-grid__item--link'):
    href = link.get('href')
    if href and 'movie' in href:
        # Extract the last part of the URL
        url_parts = href.split('/')
        # Assuming the last part of the URL is the movie title slug
        slug = url_parts[-1] if url_parts[-1] else url_parts[-2]  # Handling trailing slash
        # Converting the slug into a title (replace hyphens/underscores with spaces and capitalize)
        title = slug.replace('-', ' ').replace('_', ' ').title()
        movie_titles.append(title)

# Printing the movie titles extracted from URLs


for title in movie_titles:
    print(title)
````

## **Scrapping Genres**

````python
# Write Your Code here
import requests
from bs4 import BeautifulSoup

# Initialize a dictionary to store genres for each movie URL
movie_genres = {}

# Headers for the requests
headers = {'User-Agent': 'Mozilla/5.0'}

# Iterate through each movie URL and scrape for genres
for url in movie_urls:
    # Send an HTTP GET request to the movie URL and parse the HTML content
    movie_page = requests.get(url, headers=headers)
    movie_soup = BeautifulSoup(movie_page.text, 'html.parser')

    # Find the <h3> element with the text 'Genres'
    genres_heading = movie_soup.find('h3', class_='detail-infos__subheading', string='Genres')

    # Default text if no genres are found
    genres_text = 'No genres'

    # If the <h3> element is found, find the next <div> element with class 'detail-infos__value'
    if genres_heading:
        genres_div = genres_heading.find_next_sibling('div', class_='detail-infos__value')
        if genres_div:
            genres_text = genres_div.get_text(strip=True)

    movie_genres[url] = genres_text

# Print the results
for movie_url, genres in movie_genres.items():
    print(genres)

````

## **Scrapping IMBD Rating**

````python
# Write Your Code here
# Initialize a dictionary to store IMDb ratings for each movie URL
movie_imdb_ratings = {}

# Iterate through each movie URL and scrape for IMDb ratings
for url in movie_urls:
    try:
        # Send an HTTP GET request to the movie URL and parse the HTML content
        movie_page = requests.get(url, headers=headers)
        movie_soup = BeautifulSoup(movie_page.text, 'html.parser')

        # Find the outer span tag by locating the img tag with alt="IMDB"
        imdb_img_tag = movie_soup.find('img', alt="IMDB")
        if imdb_img_tag:
            outer_span = imdb_img_tag.parent
            if outer_span:
                # Find the nested span tag which should contain the IMDb rating
                imdb_rating_span = outer_span.find('span')
                imdb_rating = imdb_rating_span.get_text().strip()[:3] if imdb_rating_span else "Rating not found"
            else:
                imdb_rating = "Outer span tag not found"
        else:
            imdb_rating = "IMDb image tag not found"

        movie_imdb_ratings[url] = imdb_rating

    except Exception as e:
        movie_imdb_ratings[url] = f"Error occurred: {e}"

# Print the results
for movie_url, rating in movie_imdb_ratings.items():
    print(rating)


````

## **Analyzing Top Genres**

````python
# Write Your Code here
import pandas as pd

# ... (Assuming you have already defined your DataFrames Tvshow_DF and movie_DF)

# Combine the 'Genres' column from both DataFrames
all_genres = Tvshow_DF['Genres'].tolist() + movie_DF['Genres'].tolist()

# Flatten the list of lists into a single list
all_genres_flat = [genre for sublist in all_genres for genre in sublist]

# Create a DataFrame to count the occurrences of each genre
genres_count = pd.DataFrame({'Genre': all_genres_flat})
genres_count['Count'] = genres_count.groupby('Genre')['Genre'].transform('count')
genres_count = genres_count.drop_duplicates().reset_index(drop=True)

# Sort the DataFrame by count in descending order
top_genres = genres_count.sort_values(by='Count', ascending=False)

# Display the top genres and their occurrence counts
print("Top Genres:")
print(top_genres)
````
